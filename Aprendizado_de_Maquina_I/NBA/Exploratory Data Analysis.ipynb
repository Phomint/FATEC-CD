# **Exploratory Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn.apionly as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

## **Histogram**

```python
# read dataset
df = pd.read_csv('winequality-white.csv', sep=';')

# create histogram
bin_edges = np.arange(0, df['residual sugar'].max() + 1, 1)
fig = plt.hist(df['residual sugar'], bins=bin_edges)

# add plot labels
plt.xlabel('count')
plt.ylabel('residual sugar')
plt.show()
```

## **Scatterplot**

```python
# create scatterplot
fig = plt.scatter(df['pH'], df['residual sugar'])

# add plot labels
plt.xlabel('pH')
plt.ylabel('residual sugar')
plt.show()
```

## **Scatterplot Matrix**

```python
df.columns
```

```python
# create scatterplot matrix
fig = sns.pairplot(data=df[['alcohol', 'pH', 'residual sugar', 'quality']], 
                   hue='quality')

# add plot labels
plt.xlabel('pH')
plt.ylabel('residual sugar')
plt.show()
```

## **Bee Swarm Plot**

- useful for small datasets but can be slow on large datasets

```python
# create bee swarm plot
sns.swarmplot(x='quality', y='residual sugar', 
              data=df[df['quality'] < 6])
plt.show()
```

## **Boxplots**

- Distribution of data in terms of median and percentiles (median is the 50th percentile)

```python
percentiles = np.percentile(df['alcohol'], q=[25, 50, 75])
percentiles
```

```python
plt.boxplot(df['alcohol'])

plt.ylim([8, 16])
plt.ylabel('alcohol')

fig = plt.gca()
fig.axes.get_xaxis().set_ticks([])
plt.show()
```

## **Violin Plots**

```python
plt.violinplot(df['alcohol'], [0], 
               points=100, 
               bw_method='scott',
               showmeans=False, 
               showextrema=True, 
               showmedians=True)

plt.ylim([8, 16])
plt.ylabel('alcohol')
fig = plt.gca()
fig.axes.get_xaxis().set_ticks([])
plt.show()
```
